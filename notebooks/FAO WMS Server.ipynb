```python
from owslib.wms import WebMapService
```

```python
faowmsurl = 'https://io.apps.fao.org/geoserver/wms/ASIS/ASI_D/v1?service=WMS&version=1.3.0&request=GetCapabilities'

wms = WebMapService(faowmsurl, version='1.3.0')
```

```python
list(wms.contents)
```

```python
wms['ASI_D_1985-05_S2_C:ASIS:asis_asi_d_c'].title
```

```python
[op.name for op in wms.operations]
```

```python
wms.getOperationByName('GetMap').methods
```

```python
wms.getOperationByName('GetMap').formatOptions
```

```python
tz_bbox = (29.3399975929, -11.7209380022, 40.31659, -0.95)
img = wms.getmap(   layers=['ASI_D_1985-05_S2_C:ASIS:asis_asi_d_c'],

                     srs='EPSG:4326',
                     bbox=tz_bbox,
                     size=(300, 250),
                     format='image/geotiff',
                     transparent=True
                     )
out = open('a.tiff', 'wb')
out.write(img.read())
out.close()
```
