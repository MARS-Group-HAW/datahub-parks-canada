```python
def admin_level(name):
    s = re.search(r'\((L[0-9])\)', name)
    
    if not s:
        return None
    
    return s[1]

def region_name(name):
    
    if "(" in name:
        return name.split('(')[0].strip()
    
    return name


print(admin_level("Unguja (L1)"))
print(admin_level("Western"))


print(region_name("Unguja (L1)"))
print(region_name("Western"))
```
